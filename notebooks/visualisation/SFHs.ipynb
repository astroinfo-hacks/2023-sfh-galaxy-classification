```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Load binned SFHs on the redshift interval 0.5<z<1.0

```python
df=pd.read_csv('../data/binned_SFHs/binned_SFHs-7levels-JWST_z_0.5-1.0.txt',sep='\t')
df
```

## Get bins and levels names labels

```python
levels=df.columns[2:8]
bins=df.columns[8:]
print(levels,'\n',bins)
SFH_lev=df[levels].values
SFH_bin=df[bins].values
```

## Plot some SFHs

```python

for i in np.random.randint(0,91370,10):
    plt.semilogx()
    plt.hlines(SFH_lev[i],SFH_bin[i][:-1],SFH_bin[i][1:],'r')
    plt.vlines(SFH_bin[i][1:-1],SFH_lev[i][:-1],SFH_lev[i][1:],'r')
    plt.xlim(10,9000)
    plt.xlabel(r'Lookback time [Myrs]')
    plt.ylabel(r'Norm SFR [--]')
    plt.show()
```
