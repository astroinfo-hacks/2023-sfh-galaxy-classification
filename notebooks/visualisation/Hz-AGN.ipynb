## Galaxy Main Sequence from LePhare - Template fitting.

```python

redshift_lp=df_physics['ZMED_UP_sedf'].values
mass_lp=df_physics['MMED_sedf'].values
sfr_lp=df_physics['SFRMED_sedf'].values
mass_CIG=df_CIG['bayes.stellar.m_star'].values
sfr_CIG=df_CIG['bayes.sfh.sfr'].values

cond_lp=(redshift_lp>0.5)&(redshift_lp<1)

redshift=df_physics['z_true'].values
mass=df_physics['MASSBC03_true'].values
sfr=df_physics['SFR_true'].values

cond=(redshift>0.5)&(redshift<1)

plt.scatter(mass[cond],np.log10(sfr[cond]),s=0.1,c='k',label='Hz-AGN')

plt.scatter(mass_lp[cond_lp],np.log10(sfr_lp[cond_lp]),s=0.05,c='r',label='LePhare')

plt.xlabel(r'Log$_{10}$ M$_{*}$ [M$_{\odot}$]')
plt.ylabel(r'Log$_{10}$ SFR [M$_{\odot}$yr$^{-1}$]')
plt.legend()
```

## Galaxy Main Sequence from CIGALE - SED fitting.

```python

redshift_lp=df_physics['ZMED_UP_sedf'].values
mass_lp=df_physics['MMED_sedf'].values
sfr_lp=df_physics['SFRMED_sedf'].values
mass_CIG=df_CIG['bayes.stellar.m_star'].values
sfr_CIG=df_CIG['bayes.sfh.sfr'].values      #

cond_lp=(redshift_lp>0.5)&(redshift_lp<1)

redshift=df_physics['z_true'].values
mass=df_physics['MASSBC03_true'].values
sfr=df_physics['SFR_true'].values

cond=(redshift>0.5)&(redshift<1)

plt.scatter(mass[cond],np.log10(sfr[cond]),s=0.1,c='k',label='Hz-AGN')

plt.scatter(np.log10(mass_CIG),np.log10(sfr_CIG)+0.24,s=0.1,alpha=0.9,c='g',label='CIGALE')
#O.24 dex because the difference between the IMFs

plt.xlabel(r'Log$_{10}$ M$_{*}$ [M$_{\odot}$]')
plt.ylabel(r'Log$_{10}$ SFR [M$_{\odot}$yr$^{-1}$]')
plt.legend()
```

## Color-color diagram: The NUVrk 

```python


NUV=df_rest_frame['NUV_TOT_rf'].values
K=df_rest_frame['Ks_TOT_rf'].values
r=df_rest_frame['r_TOT_rf'].values
sfr=df_physics['SFR_true'].values
mass=df_physics['MASS_true'].values
sSFR=np.log10(sfr)-mass


#Inside reg -> Passive galaxies
#Outside reg -> Active galaxies
x_int=(3.75-3.2)/1.37
x_int=np.linspace(-1,x_int,10)
y_reg1=np.zeros_like(x_int)+3.75
plt.plot(x_int,y_reg1,c='k')
x_int=np.linspace(x_int[-1],1.5,10)
y_reg2=1.37*x_int + 3.2

plt.plot(x_int,y_reg2,c='k')


z=sSFR
idx = z.argsort()
x, y, z = (r-K)[idx], (NUV-r)[idx], z[idx]

sc=plt.scatter(x,y,s=0.01,c=z+9)
plt.colorbar(sc,label=r'sSFR[Gyr$^{-1}$]')
plt.xlabel(r'(r-K)$_{AB}$')
plt.ylabel(r'(NUV-r)$_{AB}$')
plt.xlim(-1,1.1)
```
